```python
import pandas as pd
from sklearn.cluster import MeanShift
from scipy.spatial import distance
from collections import Counter
```

```python
full_data = pd.read_table('checkins.dat', sep='|', skipinitialspace = True, header=[0], skiprows=[1], )
full_data.dropna(axis='rows', inplace=True)
#full_data.drop(['created_at'], axis=1)
full_data.columns = full_data.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('(', '').str.replace(')', '')
full_data.drop(['created_at', 'id', 'user_id', 'venue_id'], axis=1, inplace=True)
```

```python
data = full_data.head(100000)
```

```python
clustering = MeanShift(bandwidth=0.1, n_jobs = -1).fit(data)
```

```python
clustering.labels_
```

```python
cluster_centers = clustering.cluster_centers_.tolist()
```

```python
labels_counter = Counter(clustering.labels_)
```

```python
small_clust_list = []
for num, count in Counter(labels_counter).items():
    if count < 15:
        small_clust_list.append(num)
    
```

```python
available_clusters = []
for i in range(0, len(cluster_centers)):
    if labels_counter[i] > 15:
        available_clusters.append(cluster_centers[i])
```

```python
clust_nearest_dists = []
ccl_cords = [[33.751277, -118.188740], [25.867736, -80.324116], [51.503016, -0.075479], [52.378894, 4.885084], [39.366487, 117.036146], [-33.868457, 151.205134]]
for clust_cord in available_clusters:
    nearest_dist = -1
    for ccl_cord in ccl_cords:
        dst = distance.euclidean(clust_cord, ccl_cord)
        if nearest_dist == -1:
            nearest_dist = dst
        elif dst < nearest_dist:
            nearest_dist = dst
    clust_nearest_dists.append((dst, clust_cord))
            
        
```

```python
clust_nearest_dists.sort(key=lambda x: x[0])
```
